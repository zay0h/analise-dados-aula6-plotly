## 4. Missão 1 — Barras: Receita por Canal

A missão prática da aula pede construir um gráfico de barras para responder:
**qual canal vende mais?**  
Os slides também sugerem usar o hover para injetar métricas secundárias sem poluir a tela. fileciteturn7file0

### Tarefa
1. Agregue a `receita` por `canal_venda`
2. Inclua também uma métrica secundária, como `quantidade`
3. Construa um gráfico de barras com Plotly Express
4. Ordene do maior para o menor
5. Faça o hover mostrar mais do que apenas a receita

### Perguntas
- Qual canal lidera a receita?
- O canal líder também lidera em quantidade?
- O hover ajudou a enriquecer a leitura sem poluir a tela?


```python
# agrupar dados
tabela_canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'quantidade': 'sum'
}).reset_index()

# ordenar do maior para o menor
tabela_canal = tabela_canal.sort_values(by='receita', ascending=False)

# criar gráfico interativo
fig = px.bar(
    tabela_canal,
    x='canal_venda',
    y='receita',
    color='canal_venda',
    title="Receita por Canal de Venda",
    hover_data=['quantidade']
)

fig.show()
```

### Insight obrigatório
Escreva 2 ou 3 linhas explicando:
- quem lidera
- quem fica atrás
- o que um gestor poderia investigar a seguir


O gráfico mostra qual canal de venda lidera em geração de receita, destacando o principal responsável pelo faturamento do negócio. Também é possível identificar os canais com menor participação, que aparecem com barras menores.

Com base nisso, um gestor pode investigar se os canais que ficam atrás possuem menor volume de vendas, menor ticket médio ou se existem oportunidades de melhoria em estratégias comerciais ou de marketing nesses canais.

## 5. Missão 2 — Linhas: Sazonalidade da Receita Mensal

Os slides destacam que o gráfico de linhas, com zoom e pan, é ideal para navegar no tempo e isolar picos. Também reforçam que o eixo temporal precisa estar corretamente tipado no Pandas. fileciteturn7file0

### Tarefa
1. Converta `data_venda` em data, se necessário
2. Agregue a `receita` por mês
3. Construa um gráfico de linha interativo
4. Teste o zoom nos meses de pico
5. Observe se novembro e dezembro se destacam

### Perguntas
- Existe sazonalidade?
- Quais meses chamam atenção?
- O zoom ajudou a explorar melhor a parte final da série?


```python
# converter data para datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])

# criar coluna de mês
df['mes'] = df['data_venda'].dt.to_period('M').astype(str)

# agrupar receita por mês
receita_mes = df.groupby('mes')['receita'].sum().reset_index()

# criar gráfico interativo
fig = px.line(
    receita_mes,
    x='mes',
    y='receita',
    title="Evolução da Receita Mensal",
    markers=True
)

fig.show()
```

### Insight obrigatório
Explique:
- qual tendência aparece
- se há picos claros
- qual hipótese de negócio pode explicar o comportamento observado


O gráfico mostra a evolução da receita ao longo do tempo, indicando uma tendência de variação nas vendas entre os meses.

É possível observar alguns picos claros em determinados períodos, o que pode indicar momentos de maior demanda no mercado.

Uma possível hipótese de negócio é que esses picos estejam relacionados a datas sazonais, campanhas promocionais ou aumento do consumo em períodos específicos do ano.

## 6. Missão 3 — Dispersão: Lucro vs. Receita segmentado por Categoria

### Perguntas

```python
fig = px.scatter(
    df,
    x='receita',
    y='lucro',
    color='categoria',
    hover_data=['produto','canal_venda','uf','margem_lucro'],
    title='Relação entre Receita e Lucro por Categoria',
    opacity=0.7
)

fig.show()
```

O uso do hover permite identificar exatamente qual produto, canal ou região está associado a cada ponto, transformando o gráfico em uma ferramenta útil para investigar casos específicos dentro da operação.

## 7. Exploração espacial — Onde está concentrada a operação?

Os slides incluem mapas geográficos como resposta para perguntas espaciais como:
**onde está concentrada nossa operação?** fileciteturn7file0

### Tarefa
Usando `latitude` e `longitude`, crie um mapa interativo que ajude a visualizar a operação.

### Sugestões
- use tamanho ou cor para uma métrica relevante (como receita)
- teste o zoom do mapa
- observe se há concentração regional

### Perguntas
- Onde a operação parece mais concentrada?
- O mapa ajudou mais do que uma simples tabela por UF?


```python
fig = px.scatter_mapbox(
    df,
    lat='latitude',
    lon='longitude',
    size='receita',
    color='receita',
    hover_name='produto',
    hover_data=['canal_venda','uf','lucro'],
    zoom=3,
    height=500,
    title='Distribuição geográfica da receita'
)

fig.update_layout(mapbox_style="open-street-map")

fig.show()
```

O mapa mostra que a operação está mais concentrada em determinadas regiões, onde os pontos aparecem maiores e mais intensos.

Essas áreas representam maior geração de receita e podem indicar mercados mais fortes ou maior presença da empresa.

O mapa ajuda mais do que uma tabela por UF, pois permite visualizar rapidamente a concentração geográfica e identificar padrões espaciais de forma intuitiva.

```python
tabela_canal = df.groupby('canal_venda')['receita'].sum().reset_index()

tabela_canal = tabela_canal.sort_values(by='receita', ascending=False)

fig = px.bar(
    tabela_canal,
    x='canal_venda',
    y='receita',
    title='Receita Total por Canal de Venda',
    labels={
        'canal_venda': 'Canal de Venda',
        'receita': 'Receita Total (R$)'
    },
    color='canal_venda',
    hover_data={'receita':':,.2f'}
)

fig.update_layout(
    showlegend=False
)

fig.show()
```

Foram feitas melhorias no título, nos rótulos dos eixos e na formatação dos valores para tornar o gráfico mais claro e fácil de interpretar. Também foi removida a legenda desnecessária para reduzir o excesso de elementos visuais.
